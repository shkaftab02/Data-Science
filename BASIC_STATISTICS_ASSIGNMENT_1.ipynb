## **Descriptive Analytics for Numerical Columns**

```python
import pandas as pd
dataset = pd.read_csv('/content/sales_data_with_discounts.csv')
print(dataset.dtypes)

```

```python
numerical_data = dataset.select_dtypes(include=['float64', 'int64'])
print(numerical_data.columns)
```

```python
for column in numerical_data.columns:
    print(f"Statistics for {column}:")
    print(f"Mean: {numerical_data[column].mean()}")
    print(f"Median: {numerical_data[column].median()}")
    print(f"Mode: {numerical_data[column].mode()[0]}")
    print(f"Standard Deviation: {numerical_data[column].std()}")
    print()
```

## ***Data Visualization***

## ‚óè	**Histograms**

```python

import matplotlib.pyplot as plt

numerical_data = dataset.select_dtypes(include=['float64', 'int64'])
print(numerical_data.columns)

for column in numerical_data.columns:
    plt.figure(figsize=(8, 5))
    plt.hist(numerical_data[column], bins=20, color='blue', alpha=0.7)
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

```

# ‚óè	Boxplots

```python
import seaborn as sns

numerical_data = dataset.select_dtypes(include=['float64', 'int64'])
print(numerical_data.columns)

for column in numerical_data.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=numerical_data[column], color='skyblue')
    plt.title(f"Boxplot of {column}")
    plt.xlabel(column)
    plt.grid(True)
    plt.show()

```

# **‚óè	Bar Chart Analysis for Categorical Column**

```python
categorical_data = dataset.select_dtypes(include=['object', 'category'])
print(categorical_data.columns)


for column in categorical_data.columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(data=dataset, x=column, palette='viridis')
    plt.title(f"Bar Chart of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

## **Standardization of Numerical Variables**

```python
numerical_data = dataset.select_dtypes(include=['float64', 'int64'])
print(numerical_data.columns)

from sklearn.preprocessing import StandardScaler
import pandas as pd
scaler = StandardScaler()
standardized_data = pd.DataFrame(scaler.fit_transform(numerical_data),
                                 columns=numerical_data.columns)
data_standardized = dataset.copy()
data_standardized[numerical_data.columns] = standardized_data

print("Before Standardization:")
print(numerical_data.describe())

print("After Standardization:")
print(standardized_data.describe())

```

## **Conversion of Categorical Data into Dummy Variables**

```python
categorical_data = dataset.select_dtypes(include=['object', 'category'])
print(categorical_data.columns)

one_hot_encoded_data = pd.get_dummies(dataset, columns=categorical_data.columns, drop_first=False)

print(one_hot_encoded_data.head())

print("Original Dataset:")
print(dataset.head())

print("\nTransformed Dataset with One-Hot Encoding:")
print(one_hot_encoded_data.head())
```

## ***Conclusion ***

Key Findings:

Descriptive analytics (mean, median, standard deviation) showed central trends and data spread.
Visualizations (boxplots, histograms, bar charts) helped identify outliers, distribution patterns, and category frequencies.

Importance of Preprocessing:
